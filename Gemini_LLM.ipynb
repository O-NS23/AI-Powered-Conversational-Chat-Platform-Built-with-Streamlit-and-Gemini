```python
import google.generativeai as genai
from google.colab import userdata

# 1. Setup
my_key = userdata.get('GOOGLE_API_KEY')
genai.configure(api_key=my_key)

# 2. THE SEARCH PARTY: Find which models you can actually use
print("Searching for available brains...")
available_models = [m.name for m in genai.list_models() if 'generateContent' in m.supported_generation_methods]

if not available_models:
    print("‚ùå No models found. Check your API key!")
else:
    # Pick the first smart model found (usually gemini-1.5-flash or gemini-pro)
    chosen_model = available_models[0]
    print(f"‚úÖ Found it! Using: {chosen_model}")

    # 3. Connect and Chat
    model = genai.GenerativeModel(chosen_model)
    user_question = "How should a Data Science Fresher Resume look?"

    try:
        response = model.generate_content(user_question)
        print("-" * 30)
        print(f"AI: {response.text}")
    except Exception as e:
        print(f"Still a snag: {e}")
```

```python
%%writefile app.py
import streamlit as st
import google.generativeai as genai

# ---------- CONFIGURE GEMINI ----------
genai.configure(api_key="AIzaSyA9obsQp62dLdrZcioDhko_AzYEr15oUX4")
model = genai.GenerativeModel("gemini-1.5-flash")

st.set_page_config(page_title="Smart Chat ü§ñ")
st.title("Continuous Chat App ü§ñ")

# ---------- CHAT MEMORY ----------
if "chat" not in st.session_state:
    st.session_state.chat = []

question = st.text_input("Ask your question üëá")

col1, col2 = st.columns(2)

# ---------- GET ANSWER BUTTON ----------
with col1:
    if st.button("Get Answer"):
        if question.strip() == "":
            st.warning("Please type something first üìù")
        else:
            with st.spinner("Thinking... ü§î"):
                reply = model.generate_content(question).text

            # save conversation
            st.session_state.chat.append(("You", question))
            st.session_state.chat.append(("AI", reply))

# ---------- END CHAT ----------
with col2:
    if st.button("End Here"):
        st.session_state.chat = []
        st.success("Chat Ended üòä")

# ---------- DISPLAY CHAT ----------
st.write("### Chat History")
for sender, msg in st.session_state.chat:
    if sender == "You":
        st.markdown(f"üßë **You:** {msg}")
    else:
        st.markdown(f"ü§ñ **AI:** {msg}")
```
